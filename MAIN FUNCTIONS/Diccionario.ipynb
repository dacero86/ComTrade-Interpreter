## Funciones

```python
def carga_Siemens(n):
  rec = Comtrade()
  rec.load(n,encoding = "ISO-8859-1")
  names_columns = rec.analog_channel_ids + rec.digital_channel_ids
  list(names_columns)
  return names_columns
```

```python
def carga(n):
  rec = Comtrade()
  rec.load(n)
  names_columns = rec.analog_channel_ids + rec.digital_channel_ids
  list(names_columns)
  return names_columns
```

```python
def str_plit(diccionario):
    aux = []
    var_str = []
    for i in range(len(diccionario)):
        str_sie = str(diccionario[i]).split(":")
        aux.append(str_sie)

    for i in range(len(diccionario)):
      if (len(aux[i]) == 1):
          str_sie = str(diccionario[i]).split()
          k = len(str_sie)
          for n in range(k):
            var_str.append(str_sie[n])
      else:
          var_str.extend(aux[i])

    return var_str
```

```python
def str_plit_abb(diccionario):
    aux = []
    var_str = []
    for i in range(len(diccionario)):
        str_sie = str(diccionario[i]).split("_")
        aux.append(str_sie)

    for i in range(len(diccionario)):
      if (len(aux[i]) == 1):
          str_sie = str(diccionario[i]).split()
          k = len(str_sie)
          for n in range(k):
            var_str.append(str_sie[n])
      else:
          var_str.extend(aux[i])

    return var_str
```
