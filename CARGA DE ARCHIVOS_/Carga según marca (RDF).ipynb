```python
from comtrade import Comtrade

## Ametek
rdf_1 = '/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/RDF/Ametek/ZQ6021.cfg'
rdf_2 = '/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/RDF/Ametek/ZQ6022.cfg'
rdf_3 = '/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/RDF/Ametek/ZQ6023.cfg'
rdf_4 = '/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/RDF/Ametek/ZQ6024.cfg'

## GE Reason RPV311
rdf_5 = '/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/RDF/GE Reason RPV311/220903,173229969792,-5,CIRCO,F008,GEB,000001083333,fault.cfg'
rdf_6 = '/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/RDF/GE Reason RPV311/220903,173229970378,-5,TUNAL,F018,GEB,000001083333,fault.cfg'

rdf_n = [rdf_1,rdf_2,rdf_3,rdf_4,rdf_5,rdf_6]

#rec_abb.load(r'/content/drive/MyDrive/Análisis de eventos /Oscilografias_Fabricantes/ABB/20220903 1732_30 SFOD PP1.cfg',encoding = "ISO-8859-1")
```

Función que realiza la carga de todos de los .cfg

```python
def carga(n):
  rec = Comtrade()
  rec.load(n)
  names_columns = rec.analog_channel_ids + rec.digital_channel_ids
  list(names_columns)
  return names_columns
```

```python
names_rdf = {rdf_n:carga(rdf_n) for rdf_n in rdf_n}
```

```python
for names in names_rdf:
  print(len(names_rdf[names]))
```

```python
rdf1 = set(names_rdf[rdf_1])
rdf2 = set(names_rdf[rdf_2])
rdf3 = set(names_rdf[rdf_3])
rdf4 = set(names_rdf[rdf_4])
rdf5 = set(names_rdf[rdf_5])
rdf6 = set(names_rdf[rdf_6])
```

```python
rfd_n1 = sorted(rdf1)
rfd_n2 = sorted(rdf3)
rfd_n3 = sorted(rdf5)
rfd_n4 = sorted(rdf6)
```

Se genera un dataframe

```python
rfd_n1 = pd.DataFrame(rfd_n1,columns = ['rfd1'])
rfd_n2 = pd.DataFrame(rfd_n2,columns = ['rfd2'])
rfd_n3 = pd.DataFrame(rfd_n3,columns = ['rfd3'])
rfd_n4 = pd.DataFrame(rfd_n4,columns = ['rfd4'])
```

```python
rfd = pd.concat([rfd_n1, rfd_n2,rfd_n3,rfd_n4],axis = 1)
```
