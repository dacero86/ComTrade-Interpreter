## Funciones para el procesamiento

```python
def carga(n,encoding):
  rec = Comtrade()
  rec.load(n,encoding = encoding)
  names_columns = rec.analog_channel_ids + rec.digital_channel_ids
  list(names_columns)
  return names_columns
```

```python
def str_plit(diccionario):
    aux = []
    var_str = []
    for i in range(len(diccionario)):
        str_sie = str(diccionario[i]).split(":")
        aux.append(str_sie)

    for i in range(len(diccionario)):
      if (len(aux[i]) == 1):
          str_sie = str(diccionario[i]).split()
          k = len(str_sie)
          for n in range(k):
            var_str.append(str_sie[n])
      else:
          var_str.extend(aux[i])

    return var_str
```

```python
def str_plit_abb(diccionario):
    aux = []
    var_str = []
    for i in range(len(diccionario)):
        str_sie = str(diccionario[i]).split("_")
        aux.append(str_sie)

    for i in range(len(diccionario)):
      if (len(aux[i]) == 1):
          str_sie = str(diccionario[i]).split()
          k = len(str_sie)
          for n in range(k):
            var_str.append(str_sie[n])
      else:
          var_str.extend(aux[i])

    return var_str
```

## Funciones para la traducción

```python
def var_input(names,split):
    input_names = []
    for i in names:
      p = (i.split(split))
      if (len(p)==1):
        input_names.append(p[0].split())
      else:
        input_names.append(p)

    return input_names
```

```python
def search_dicc(input,diccionario):
    traduccion = []
    without_translation = []
    for i in input:
      aux_var = diccionario.get(i)
      if (aux_var == None):
          try:
              aux_var = diccionario.get(int(i))
          except ValueError:
              aux_var = i
              without_translation.append([i,input])

              ## Función que pregunte por las que no tienen traducción, y las agrege a la base de datos
      traduccion.append(aux_var)

    translation  = " ".join(traduccion)

    return translation
```
